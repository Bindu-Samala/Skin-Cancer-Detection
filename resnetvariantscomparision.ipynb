```python
import pandas as pd
import numpy as np
import os
import tensorflow as tf
import time
import psutil
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc, precision_score, recall_score, f1_score
from sklearn.preprocessing import label_binarize
from sklearn.model_selection import train_test_split
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import SMOTE
from tensorflow.keras import layers, models
from tensorflow.keras.preprocessing import image
from PIL import Image

# 1. Load the metadata
metadata_path = '/kaggle/input/skin-cancer-mnist-ham10000/HAM10000_metadata.csv'
metadata = pd.read_csv(metadata_path)

# Extract image file names and labels
image_files = metadata['image_id'].values
labels = metadata['dx'].values

# Ensure labels are categorical
label_map = {label: idx for idx, label in enumerate(np.unique(labels))}
y = np.array([label_map[label] for label in labels])

# Set image directories
image_dir_part1 = '/kaggle/input/skin-cancer-mnist-ham10000/HAM10000_images_part_1/'
image_dir_part2 = '/kaggle/input/skin-cancer-mnist-ham10000/HAM10000_images_part_2/'

# Function to load and resize images
def load_and_resize_image(image_path, target_size=(112, 112)):
    img = Image.open(image_path)
    img = img.convert('RGB')
    img = img.resize(target_size)
    img_array = np.array(img)
    return img_array

# Load and resize images
x = []
for img_id in image_files:
    if os.path.exists(os.path.join(image_dir_part1, f'{img_id}.jpg')):
        img_path = os.path.join(image_dir_part1, f'{img_id}.jpg')
    elif os.path.exists(os.path.join(image_dir_part2, f'{img_id}.jpg')):
        img_path = os.path.join(image_dir_part2, f'{img_id}.jpg')
    else:
        continue
    
    img_array = load_and_resize_image(img_path)
    x.append(img_array)

x = np.array(x)
x = (x - np.mean(x)) / np.std(x)  # Normalize images

x_flattened = x.reshape(x.shape[0], -1)

# Resampling
majority_class_size = pd.Series(y).value_counts().max()
majority_class_label = pd.Series(y).value_counts().idxmax()
target_size_majority = majority_class_size // 2

# Undersampling
undersample = RandomUnderSampler(sampling_strategy={majority_class_label: target_size_majority})
x_undersampled, y_undersampled = undersample.fit_resample(x_flattened, y)

# Apply SMOTE to minority classes
minority_classes = pd.Series(y_undersampled).value_counts()[pd.Series(y_undersampled).value_counts() < target_size_majority].index
sampling_strategy = {cls: target_size_majority for cls in minority_classes}
smote = SMOTE(sampling_strategy=sampling_strategy, random_state=42, k_neighbors=10)
x_resampled_flattened, y_resampled = smote.fit_resample(x_undersampled, y_undersampled)

# Reshape back to original 4D shape for CNN
x_resampled = x_resampled_flattened.reshape(-1, 112, 112, 3)

# Split data into training, validation, and test sets
X_train, X_temp, Y_train, Y_temp = train_test_split(x_resampled, y_resampled, test_size=0.3, random_state=42)
X_val, X_test, Y_val, Y_test = train_test_split(X_temp, Y_temp, test_size=0.5, random_state=42)

# Define model
input_shape = (112, 112, 3)
num_classes = len(np.unique(y_resampled))

def create_custom_resnet_model(input_shape, num_classes):
    inputs = layers.Input(shape=input_shape)
    x = layers.Conv2D(64, (3, 3), padding='same', activation='relu')(inputs)
    x = layers.BatchNormalization()(x)
    x = layers.MaxPooling2D(pool_size=(2, 2))(x)

    def resnet_block(x, filters):
        shortcut = x
        x = layers.Conv2D(filters, (3, 3), padding='same', activation='relu')(x)
        x = layers.BatchNormalization()(x)
        x = layers.Conv2D(filters, (3, 3), padding='same')(x)
        x = layers.BatchNormalization()(x)
        if shortcut.shape[-1] != filters:
            shortcut = layers.Conv2D(filters, (1, 1), padding='same')(shortcut)
            shortcut = layers.BatchNormalization()(shortcut)
        x = layers.add([x, shortcut])
        x = layers.Activation('relu')(x)
        return x

    x = resnet_block(x, 64)
    x = layers.MaxPooling2D(pool_size=(2, 2))(x)
    x = resnet_block(x, 128)
    x = resnet_block(x, 128)
    x = layers.MaxPooling2D(pool_size=(2, 2))(x)
    x = resnet_block(x, 256)
    x = resnet_block(x, 512)
    x = layers.MaxPooling2D(pool_size=(2, 2))(x)

    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dense(256, activation='relu')(x)
    x = layers.Dropout(0.5)(x)
    outputs = layers.Dense(num_classes, activation='softmax')(x)
    model = models.Model(inputs=inputs, outputs=outputs)
    return model

reduce_lr = tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=1e-6)
callback = tf.keras.callbacks.ModelCheckpoint(
    filepath='best_model.keras',
    monitor='val_accuracy',  # Monitor validation accuracy for TensorFlow 2.x
    mode='max',
    save_best_only=True,     # Only save the model if the monitored metric improves
    verbose=1                # Print a message when saving the model
)    

# Compile the model
model = create_custom_resnet_model(input_shape, num_classes)
model.compile(optimizer=tf.keras.optimizers.RMSprop(learning_rate=0.0001, rho=0.95),
              loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Model summary and parameters
model.summary()
total_params = model.count_params()

# Measure latency and throughput
def measure_latency_throughput(model, X_test, batch_size=32):
    start_time = time.time()
    _ = model.predict(X_test[:batch_size])
    latency = time.time() - start_time
    throughput = batch_size / latency
    return latency, throughput

latency, throughput = measure_latency_throughput(model, X_test)

# Measure memory usage
def get_memory_usage():
    process = psutil.Process(os.getpid())
    memory_info = process.memory_info()
    return memory_info.rss / (1024 * 1024)

memory_before = get_memory_usage()

# Track training time per epoch
epoch_times = []

# Custom callback to measure epoch times
class EpochTimeCallback(tf.keras.callbacks.Callback):
    def on_epoch_begin(self, epoch, logs=None):
        self.epoch_start_time = time.time()

    def on_epoch_end(self, epoch, logs=None):
        epoch_duration = time.time() - self.epoch_start_time
        epoch_times.append(epoch_duration)

# Train the model with custom callback to measure epoch times
epoch_time_callback = EpochTimeCallback()

# Measure training time
start_time = time.time()  # Start time before training

# Train the model with custom callback to measure epoch times
history = model.fit(X_train, Y_train, validation_data=(X_val, Y_val), epochs=50, batch_size=32, 
                    callbacks=[callback, reduce_lr, epoch_time_callback])

end_time = time.time()  # End time after training

# Calculate total training time
total_training_time = end_time - start_time  # Total training time in seconds

model_depth = len(model.layers)
model.save('model_after_training.h5')  # Save the model
model_size = os.path.getsize('model_after_training.h5') / (1024 * 1024)

# Convert training time to hours, minutes, and seconds for easier readability
hours, remainder = divmod(total_training_time, 3600)
minutes, seconds = divmod(remainder, 60)
total_training_time_str = f"{int(hours)}h {int(minutes)}m {int(seconds)}s"

# Calculate average training time per epoch
avg_epoch_time = np.mean(epoch_times)

# Evaluate model
test_loss, test_accuracy = model.evaluate(X_test, Y_test)

# Classification report
y_pred = model.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)

# Compute additional metrics
precision = precision_score(Y_test, y_pred_classes, average='weighted')
recall = recall_score(Y_test, y_pred_classes, average='weighted')
f1 = f1_score(Y_test, y_pred_classes, average='weighted')

# Top-1 and Top-5 accuracy
def top_k_accuracy(y_true, y_pred, k=1):
    top_k_preds = np.argsort(y_pred, axis=1)[:, -k:]
    correct = np.sum([y_true[i] in top_k_preds[i] for i in range(len(y_true))])
    return correct / len(y_true)

top_1_accuracy = top_k_accuracy(Y_test, y_pred, k=1)
top_5_accuracy = top_k_accuracy(Y_test, y_pred, k=3)

# Collect all metrics into a table
metrics = {
    'Metric': ['Test Accuracy', 'Test Loss', 'Total Parameters', 'Latency (seconds)', 'Throughput (images/sec)',
               'Memory Usage Before Training (MB)', 'Memory Usage After Training (MB)', 'Precision', 'Recall', 'F1 Score', 'Top-1 Accuracy', 'Top-5 Accuracy', 'Avg Training Time per Epoch (s)', 'Model Depth', 'Model Size After Training (MB)', 'Total Training Time'],
    'Value': [
        test_accuracy * 100,  # Test Accuracy in percentage
        test_loss,  # Test Loss
        total_params,  # Total Parameters
        latency,  # Latency (seconds)
        throughput,  # Throughput (images/sec)
        memory_before,  # Memory Usage Before Training (MB)
        get_memory_usage(),  # Memory Usage After Training (MB)
        precision,  # Precision
        recall,  # Recall
        f1,  # F1 Score
        top_1_accuracy * 100,  # Top-1 Accuracy in percentage
        top_5_accuracy * 100,  # Top-5 Accuracy in percentage
        avg_epoch_time,  # Avg Training Time per Epoch
        model_depth,  # Model Depth
        model_size,  # Model Size After Training (MB)
        total_training_time_str  # Total Training Time
    ]
}

# Convert to DataFrame for better readability
metrics_df = pd.DataFrame(metrics)
print(metrics_df)

# Classification report
class_names = list(label_map.keys())
print("\nClassification Report:")
print(classification_report(Y_test, y_pred_classes,target_names=class_names))

# Plot the loss and accuracy curves
plt.figure(figsize=(12, 6))

# Loss curve
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Loss Curve')
plt.legend()

# Accuracy curve
plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Accuracy Curve')
plt.legend()

plt.tight_layout()
plt.show()


# Plot confusion matrix
conf_matrix = confusion_matrix(Y_test, y_pred_classes)
plt.figure(figsize=(10, 8))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=label_map.keys(), yticklabels=label_map.keys())
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()

# Binarize the true labels for ROC calculation
y_test_bin = label_binarize(Y_test, classes=np.unique(Y_test))
y_pred_bin = model.predict(X_test)

# Number of classes
n_classes = len(np.unique(Y_test))

# Plot ROC curve for each class
plt.figure(figsize=(6, 5))

# For each class, calculate and plot the ROC curve
for i in range(n_classes):
    fpr, tpr, _ = roc_curve(y_test_bin[:, i], y_pred_bin[:, i])
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, lw=2, label=f'Class {i} - {list(label_map.keys())[i]} (AUC = {roc_auc:.2f})')

# Plot diagonal line (random guessing)
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')

# Set plot labels and title
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) for All Classes')

# Show legend
plt.legend(loc='lower right')
plt.show()

```

```python
import pandas as pd
import numpy as np
import os
import tensorflow as tf
import time
import psutil
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc, precision_score, recall_score, f1_score
from sklearn.preprocessing import label_binarize
from tensorflow.keras import layers, models
from tensorflow.keras.preprocessing import image
from PIL import Image
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split

# 1. Load the metadata
metadata_path = '/kaggle/input/skin-cancer-mnist-ham10000/HAM10000_metadata.csv'
metadata = pd.read_csv(metadata_path)

# Extract image file names and labels
image_files = metadata['image_id'].values
labels = metadata['dx'].values

# Ensure labels are categorical
label_map = {label: idx for idx, label in enumerate(np.unique(labels))}
y = np.array([label_map[label] for label in labels])

# Set image directories
image_dir_part1 = '/kaggle/input/skin-cancer-mnist-ham10000/HAM10000_images_part_1/'
image_dir_part2 = '/kaggle/input/skin-cancer-mnist-ham10000/HAM10000_images_part_2/'

# Function to load and resize images
def load_and_resize_image(image_path, target_size=(112, 112)):
    img = Image.open(image_path)
    img = img.convert('RGB')
    img = img.resize(target_size)
    img_array = np.array(img)
    return img_array

# Load and resize images
x = []
for img_id in image_files:
    if os.path.exists(os.path.join(image_dir_part1, f'{img_id}.jpg')):
        img_path = os.path.join(image_dir_part1, f'{img_id}.jpg')
    elif os.path.exists(os.path.join(image_dir_part2, f'{img_id}.jpg')):
        img_path = os.path.join(image_dir_part2, f'{img_id}.jpg')
    else:
        continue
    
    img_array = load_and_resize_image(img_path)
    x.append(img_array)

x = np.array(x)
x = (x - np.mean(x)) / np.std(x)  # Normalize images

x_flattened = x.reshape(x.shape[0], -1)

# Resampling
majority_class_size = pd.Series(y).value_counts().max()
majority_class_label = pd.Series(y).value_counts().idxmax()
target_size_majority = majority_class_size // 2

# Undersampling
undersample = RandomUnderSampler(sampling_strategy={majority_class_label: target_size_majority})
x_undersampled, y_undersampled = undersample.fit_resample(x_flattened, y)

# Apply SMOTE to minority classes
minority_classes = pd.Series(y_undersampled).value_counts()[pd.Series(y_undersampled).value_counts() < target_size_majority].index
sampling_strategy = {cls: target_size_majority for cls in minority_classes}
smote = SMOTE(sampling_strategy=sampling_strategy, random_state=42, k_neighbors=10)
x_resampled_flattened, y_resampled = smote.fit_resample(x_undersampled, y_undersampled)

# Reshape back to original 4D shape for CNN
x_resampled = x_resampled_flattened.reshape(-1, 112, 112, 3)

# Split data into training, validation, and test sets
X_train, X_temp, Y_train, Y_temp = train_test_split(x_resampled, y_resampled, test_size=0.3, random_state=42)
X_val, X_test, Y_val, Y_test = train_test_split(X_temp, Y_temp, test_size=0.5, random_state=42)
print("SMOTE DONE")
# ResNet Block Definition
def resnet_block(x, filters, kernel_size=3, stride=1):
    """
    A basic ResNet block with skip connection.
    Arguments:
    - x: Input tensor
    - filters: Number of filters for the convolution layer
    - kernel_size: Size of the convolution kernel
    - stride: Stride for convolution layer
    """
    shortcut = x

    # First convolutional layer
    x = layers.Conv2D(filters, (kernel_size, kernel_size), strides=stride, padding='same')(x)
    x = layers.BatchNormalization()(x)
    x = layers.ReLU()(x)

    # Second convolutional layer
    x = layers.Conv2D(filters, (kernel_size, kernel_size), strides=stride, padding='same')(x)
    x = layers.BatchNormalization()(x)

    # Adjust the shortcut if input and output channels do not match
    if shortcut.shape[-1] != filters:
        shortcut = layers.Conv2D(filters, (1, 1), strides=stride, padding='same')(shortcut)
        shortcut = layers.BatchNormalization()(shortcut)
    
    # Adjust spatial dimensions if necessary
    if shortcut.shape[1] != x.shape[1] or shortcut.shape[2] != x.shape[2]:
        shortcut = layers.Conv2D(filters, (1, 1), strides=stride, padding='same')(shortcut)

    x = layers.add([x, shortcut])
    x = layers.ReLU()(x)
    
    return x

# Build ResNet-34 model
def create_resnet34(input_shape, num_classes):
    inputs = layers.Input(shape=input_shape)
    
    # Initial Convolution Layer
    x = layers.Conv2D(64, (7, 7), strides=2, padding='same')(inputs)
    x = layers.BatchNormalization()(x)
    x = layers.ReLU()(x)
    x = layers.MaxPooling2D(pool_size=(3, 3), strides=2, padding='same')(x)

    # Define ResNet blocks (ResNet-34 has 4 stages)
    # Stage 1 (3 blocks)
    for _ in range(3):
        x = resnet_block(x, 64)
    
    # Stage 2 (4 blocks)
    x = resnet_block(x, 128, stride=2)
    for _ in range(3):
        x = resnet_block(x, 128)

    # Stage 3 (6 blocks)
    x = resnet_block(x, 256, stride=2)
    for _ in range(5):
        x = resnet_block(x, 256)

    # Stage 4 (3 blocks)
    x = resnet_block(x, 512, stride=2)
    for _ in range(2):
        x = resnet_block(x, 512)

    # Global Average Pooling
    x = layers.GlobalAveragePooling2D()(x)
    
    # Fully connected layer
    x = layers.Dense(512, activation='relu')(x)
    x = layers.Dropout(0.5)(x)
    
    # Output layer with softmax activation
    outputs = layers.Dense(num_classes, activation='softmax')(x)

    # Create the model
    model = models.Model(inputs=inputs, outputs=outputs)
    return model

# Callbacks
reduce_lr = tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=1e-6)
callback = tf.keras.callbacks.ModelCheckpoint(
    filepath='best_model.keras',
    monitor='val_accuracy',  # Monitor validation accuracy for TensorFlow 2.x
    mode='max',
    save_best_only=True,     # Only save the model if the monitored metric improves
    verbose=1                # Print a message when saving the model
)    

# Compile the model
model = create_resnet34(input_shape=(112, 112, 3), num_classes=len(np.unique(y_resampled)))
model.compile(optimizer=tf.keras.optimizers.RMSprop(learning_rate=0.0001, rho=0.95),
              loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Model summary and parameters
model.summary()
total_params = model.count_params()

# Measure latency and throughput
def measure_latency_throughput(model, X_test, batch_size=32):
    start_time = time.time()
    _ = model.predict(X_test[:batch_size])
    latency = time.time() - start_time
    throughput = batch_size / latency
    return latency, throughput

latency, throughput = measure_latency_throughput(model, X_test)

# Measure memory usage
def get_memory_usage():
    process = psutil.Process(os.getpid())
    memory_info = process.memory_info()
    return memory_info.rss / (1024 * 1024)

memory_before = get_memory_usage()

# Train the model
history = model.fit(X_train, Y_train, epochs=50, batch_size=32, validation_data=(X_val, Y_val), callbacks=[callback, reduce_lr])

# Evaluate the model
test_loss, test_accuracy = model.evaluate(X_test, Y_test)
y_pred = model.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)

# Compute additional metrics
precision = precision_score(Y_test, y_pred_classes, average='weighted')
recall = recall_score(Y_test, y_pred_classes, average='weighted')
f1 = f1_score(Y_test, y_pred_classes, average='weighted')

# Top-k accuracy function
def top_k_accuracy(y_true, y_pred, k=1):
    top_k_preds = np.argsort(y_pred, axis=1)[:, -k:]
    correct = np.array([1 if y_true[i] in top_k_preds[i] else 0 for i in range(len(y_true))])
    return np.mean(correct)

top1_accuracy = top_k_accuracy(Y_test, y_pred, k=1)
top5_accuracy = top_k_accuracy(Y_test, y_pred, k=3)

# Confusion matrix
cm = confusion_matrix(Y_test, y_pred_classes)

# Memory usage after training
memory_after = get_memory_usage()

# Print all metrics
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")
print(f"Test Loss: {test_loss:.2f}")
print(f"Total Parameters: {total_params:,}")
print(f"Latency (seconds): {latency:.2f}")
print(f"Throughput (images/sec): {throughput:.2f}")
print(f"Memory Usage Before Training (MB): {memory_before:.2f}")
print(f"Memory Usage After Training (MB): {memory_after:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1 Score: {f1:.2f}")
print(f"Top-1 Accuracy: {top1_accuracy * 100:.2f}%")
print(f"Top-5 Accuracy: {top5_accuracy * 100:.2f}%")

# Plot confusion matrix
plt.figure(figsize=(10, 7))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=np.unique(Y_test), yticklabels=np.unique(Y_test))
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()

# ROC curve and AUC
y_test_bin = label_binarize(Y_test, classes=np.unique(Y_test))
fpr, tpr, thresholds = roc_curve(y_test_bin.ravel(), y_pred.ravel())
roc_auc = auc(fpr, tpr)

# Plot ROC curve
plt.figure(figsize=(8, 8))
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc='lower right')
plt.show()
```

```python
import pandas as pd
import numpy as np
import os
import tensorflow as tf
import time
import psutil
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc, precision_score, recall_score, f1_score
from sklearn.preprocessing import label_binarize
from tensorflow.keras import layers, models
from tensorflow.keras.preprocessing import image
from PIL import Image
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.applications.resnet50 import preprocess_input

# 1. Load the metadata
metadata_path = '/kaggle/input/skin-cancer-mnist-ham10000/HAM10000_metadata.csv'
metadata = pd.read_csv(metadata_path)

# Extract image file names and labels
image_files = metadata['image_id'].values
labels = metadata['dx'].values

# Ensure labels are categorical
label_map = {label: idx for idx, label in enumerate(np.unique(labels))}
y = np.array([label_map[label] for label in labels])

# Set image directories
image_dir_part1 = '/kaggle/input/skin-cancer-mnist-ham10000/HAM10000_images_part_1/'
image_dir_part2 = '/kaggle/input/skin-cancer-mnist-ham10000/HAM10000_images_part_2/'

# Function to load and resize images
def load_and_resize_image(image_path, target_size=(224, 224)):
    img = Image.open(image_path)
    img = img.convert('RGB')
    img = img.resize(target_size)
    img_array = np.array(img)
    return img_array

# Load and resize images
x = []
for img_id in image_files:
    if os.path.exists(os.path.join(image_dir_part1, f'{img_id}.jpg')):
        img_path = os.path.join(image_dir_part1, f'{img_id}.jpg')
    elif os.path.exists(os.path.join(image_dir_part2, f'{img_id}.jpg')):
        img_path = os.path.join(image_dir_part2, f'{img_id}.jpg')
    else:
        continue
    
    img_array = load_and_resize_image(img_path)
    x.append(img_array)

x = np.array(x)
x = preprocess_input(x)  # Preprocess images for ResNet50

# Resampling
majority_class_size = pd.Series(y).value_counts().max()
majority_class_label = pd.Series(y).value_counts().idxmax()
target_size_majority = majority_class_size // 2

# Undersampling
undersample = RandomUnderSampler(sampling_strategy={majority_class_label: target_size_majority})
x_undersampled, y_undersampled = undersample.fit_resample(x.reshape(x.shape[0], -1), y)

# Apply SMOTE to minority classes
minority_classes = pd.Series(y_undersampled).value_counts()[pd.Series(y_undersampled).value_counts() < target_size_majority].index
sampling_strategy = {cls: target_size_majority for cls in minority_classes}
smote = SMOTE(sampling_strategy=sampling_strategy, random_state=42, k_neighbors=10)
x_resampled_flattened, y_resampled = smote.fit_resample(x_undersampled, y_undersampled)

# Reshape back to original 4D shape for CNN
x_resampled = x_resampled_flattened.reshape(-1, 224, 224, 3)

# Split data into training, validation, and test sets
X_train, X_temp, Y_train, Y_temp = train_test_split(x_resampled, y_resampled, test_size=0.3, random_state=42)
X_val, X_test, Y_val, Y_test = train_test_split(X_temp, Y_temp, test_size=0.5, random_state=42)
print("SMOTE DONE")

# Load pre-trained ResNet50 model
base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# Freeze the base model layers
base_model.trainable = False

# Build the final model by adding custom layers on top
model = models.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dense(512, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(len(np.unique(y_resampled)), activation='softmax')
])

# Compile the model
model.compile(optimizer=tf.keras.optimizers.RMSprop(learning_rate=0.0001, rho=0.95),
              loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Model summary and parameters
model.summary()
total_params = model.count_params()

# Measure latency and throughput
def measure_latency_throughput(model, X_test, batch_size=32):
    start_time = time.time()
    _ = model.predict(X_test[:batch_size])
    latency = time.time() - start_time
    throughput = batch_size / latency
    return latency, throughput

latency, throughput = measure_latency_throughput(model, X_test)

# Measure memory usage
def get_memory_usage():
    process = psutil.Process(os.getpid())
    memory_info = process.memory_info()
    return memory_info.rss / (1024 * 1024)

memory_before = get_memory_usage()

# Train the model
history = model.fit(X_train, Y_train, epochs=50, batch_size=32, validation_data=(X_val, Y_val), callbacks=[callback, reduce_lr])

# Evaluate the model
test_loss, test_accuracy = model.evaluate(X_test, Y_test)
y_pred = model.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)

# Compute additional metrics
precision = precision_score(Y_test, y_pred_classes, average='weighted')
recall = recall_score(Y_test, y_pred_classes, average='weighted')
f1 = f1_score(Y_test, y_pred_classes, average='weighted')

# Top-k accuracy function
def top_k_accuracy(y_true, y_pred, k=1):
    top_k_preds = np.argsort(y_pred, axis=1)[:, -k:]
    correct = np.array([1 if y_true[i] in top_k_preds[i] else 0 for i in range(len(y_true))])
    return np.mean(correct)

top1_accuracy = top_k_accuracy(Y_test, y_pred, k=1)
top5_accuracy = top_k_accuracy(Y_test, y_pred, k=5)

# Confusion matrix
cm = confusion_matrix(Y_test, y_pred_classes)

# Memory usage after training
memory_after = get_memory_usage()

# Print all metrics
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")
print(f"Test Loss: {test_loss:.2f}")
print(f"Total Parameters: {total_params:,}")
print(f"Latency (seconds): {latency:.2f}")
print(f"Throughput (images/sec): {throughput:.2f}")
print(f"Memory Usage Before Training (MB): {memory_before:.2f}")
print(f"Memory Usage After Training (MB): {memory_after:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1 Score: {f1:.2f}")
print(f"Top-1 Accuracy: {top1_accuracy * 100:.2f}%")
print(f"Top-5 Accuracy: {top5_accuracy * 100:.2f}%")

# Plot confusion matrix
plt.figure(figsize=(10, 7))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=np.unique(Y_test), yticklabels=np.unique(Y_test))
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
# Plot accuracy
plt.figure(figsize=(12, 6))

# Plot training accuracy
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

# Plot loss
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

# Show plots
plt.tight_layout()
plt.show()

# Binarize the true labels for ROC calculation
y_test_bin = label_binarize(Y_test, classes=np.unique(Y_test))
y_pred_bin = model.predict(X_test)

# Number of classes
n_classes = len(np.unique(Y_test))

# Plot ROC curve for each class
plt.figure(figsize=(6, 5))

# For each class, calculate and plot the ROC curve
for i in range(n_classes):
    fpr, tpr, _ = roc_curve(y_test_bin[:, i], y_pred_bin[:, i])
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, lw=2, label=f'Class {i} - {list(label_map.keys())[i]} (AUC = {roc_auc:.2f})')

# Plot diagonal line (random guessing)
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')

# Set plot labels and title
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) for All Classes')
# Show legend
plt.legend(loc='lower right')
plt.show()
```
